```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Number of points
n_points = 500

# Generate random points in [-1, 1] x [-1, 1]
X = np.random.uniform(-1, 1, (n_points, 2))
x1, x2 = X[:, 0], X[:, 1]

# XOR labeling: orange if x1*x2 > 0, blue otherwise
y = (x1 * x2 > 0).astype(int)  # 1 for orange, 0 for blue

# Optional: add more features
features = np.column_stack([
    x1,
    x2,
])

plt.plot(x1[y == 0], x2[y == 0], 'bo', label='Class 0 (Blue)')
plt.plot(x1[y == 1], x2[y == 1], 'ro', label='Class 1 (Red)')
plt.title('XOR Problem Visualization')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
```
